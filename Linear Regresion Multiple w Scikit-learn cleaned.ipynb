```python
# Make a dataframe from the variables to use

variables_reg = df.filter(['OverallCond', 'HeatingQC', 'GrLivArea', 'FullBath', 
                        'BedroomAbvGr', 'KitchenQual', 'GarageCars',
                        'SalePrice'])
variables_reg
```

#### Change the data from object to int , the Regresion model requires numeric

HeatingQC: Heating quality and condition

      5 Ex	Excellent
      4 Gd	Good
      3 TA	Average/Typical
      2 Fa	Fair
      1 Po	Poor


```python
#Change the data from object to int , the Regresion model requires numeric

# change_HeatingQC

def change_HeatingQC(type):
    if type == "Ex" :
        return 5
    elif type == "Gd":
        return 4
    elif type == "TA":
        return 3
    elif type == "Fa":
        return 2
    elif type == "Po":
        return 1
    
variables_reg['HeatingQC'] = variables_reg['HeatingQC'].apply(change_HeatingQC)
variables_reg['HeatingQC'].head()
```

#### Change the data from object to int , the Regresion model requires numeric

KitchenQual: Kitchen quality

      5 Ex	Excellent
      4 Gd	Good
      3 TA	Typical/Average
      2 Fa	Fair
      1 Po	Poor


```python
#Change the data from object to int , the Regresion model requires numeric

# change_KitchenQual

def change_KitchenQual(type):
    if type == "Ex" :
        return 5
    elif type == "Gd":
        return 4
    elif type == "TA":
        return 3
    elif type == "Fa":
        return 2
    elif type == "Po":
        return 1
    
variables_reg['KitchenQual'] = variables_reg['KitchenQual'].apply(change_KitchenQual)
variables_reg['KitchenQual'].head()
```

### DROP MISSING VALUES

```python
# Take out the rows with missing values

variables_reg = variables_reg.dropna(subset=['OverallCond', 'HeatingQC', 'GrLivArea', 'FullBath', 
                        'BedroomAbvGr', 'KitchenQual', 'GarageCars',
                        'SalePrice'])

```

```python
## check if rows were droped

variables_reg
```

```python
### Convert square feet to square meter and USD TO MXN

variables_reg['GrLivArea'] = variables_reg['GrLivArea'].apply(lambda num : num / 10.764)

variables_reg['SalePrice'] = variables_reg['SalePrice'].apply(lambda num : num * 19.87)
```

```python
## Check the final dataframe
variables_reg
```

## Multiple Linear Regresion using Scikit-learn

Independent and Dependent variables

```python
#separate the other attributes from the predicting attribute
x = variables_reg.drop('SalePrice',axis=1)
#separte the predicting attribute into Y for model training 
y = variables_reg['SalePrice']
```

Splitting Data

```python
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

Applying model

```python
# importing module
from sklearn.linear_model import LinearRegression
# creating an object of LinearRegression class
LR = LinearRegression()
# fitting the training data
LR.fit(x_train,y_train)
```

Model is ready, test it with x_test 

```python
y_prediction =  LR.predict(x_test)
y_prediction
```

 Calculate the accuracy of our model

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print('r2 socre is ',score)
print('mean_sqrd_error is==',mean_squared_error(y_test,y_prediction))
print('root_mean_squared error of is==',np.sqrt(mean_squared_error(y_test,y_prediction)))
```
