## Create plot Living Space vs Sale Price

### with linear regresion line

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Open file
file_path = 'data_for_page2.csv'
df = pd.read_csv(file_path)
df 
```

```python
## Add first 50 to a dataframe
df2= df.head(50)
```

```python
# Create x variable
x =df2['GrLivArea']
x
```

```python
# Create y variable
y= df2['SalePrice']
y
```

```python
## Generate graph

# Variables x and y

# create scatter plot
plt.plot(x, y, 'o')

#m = slope, b=intercept
m, b = np.polyfit(x, y, 1)

#add line of best fit
plt.plot(x, m*x + b)

# Add title and axis names
plt.title('Living Space vs Sale Price')
plt.xlabel('Living Space')
plt.ylabel('Sale Price')
```
