## Perform Basic Data Cleaning

```python
# Make a dataframe from the variables to use

variables_reg = df.filter(['OverallCond', 'HeatingQC', 'GrLivArea', 'FullBath', 
                        'BedroomAbvGr', 'KitchenQual', 'GarageCars',
                        'SalePrice'])
variables_reg
```

```python
variables_reg.dtypes
```

#### Change the data from object to int , the Regresion model requires numeric

HeatingQC: Heating quality and condition

      5 Ex	Excellent
      4 Gd	Good
      3 TA	Average/Typical
      2 Fa	Fair
      1 Po	Poor


```python
#Change the data from object to int , the Regresion model requires numeric

# change_HeatingQC

def change_HeatingQC(type):
    if type == "Ex" :
        return 5
    elif type == "Gd":
        return 4
    elif type == "TA":
        return 3
    elif type == "Fa":
        return 2
    elif type == "Po":
        return 1
    
variables_reg['HeatingQC'] = variables_reg['HeatingQC'].apply(change_HeatingQC)
variables_reg['HeatingQC'].head()
```

#### Change the data from object to int , the Regresion model requires numeric

KitchenQual: Kitchen quality

      5 Ex	Excellent
      4 Gd	Good
      3 TA	Typical/Average
      2 Fa	Fair
      1 Po	Poor


```python
#Change the data from object to int , the Regresion model requires numeric

# change_KitchenQual

def change_KitchenQual(type):
    if type == "Ex" :
        return 5
    elif type == "Gd":
        return 4
    elif type == "TA":
        return 3
    elif type == "Fa":
        return 2
    elif type == "Po":
        return 1
    
variables_reg['KitchenQual'] = variables_reg['KitchenQual'].apply(change_KitchenQual)
variables_reg['KitchenQual'].head()
```

### DROP MISSING VALUES

```python
# Take out the rows with missing values

variables_reg = variables_reg.dropna(subset=['OverallCond', 'HeatingQC', 'GrLivArea', 'FullBath', 
                        'BedroomAbvGr', 'KitchenQual', 'GarageCars',
                        'SalePrice'])

```

```python
## check if rows were droped

variables_reg
```

```python
### Convert square feet to square meter and USD TO MXN

variables_reg['GrLivArea'] = variables_reg['GrLivArea'].apply(lambda num : num / 10.764)

variables_reg['SalePrice'] = variables_reg['SalePrice'].apply(lambda num : num * 19.87)
```
